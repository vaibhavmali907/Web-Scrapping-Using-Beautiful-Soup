```python
from bs4 import BeautifulSoup as bs
import urllib.request
import requests
import re
```

```python
link="https://www.oneill.com/fr/fr/t-shirts?cgid=men-tees&prefn1=published&prefv1=true&prefn2=seasons&prefv2=NOS%7c182%7c185%7c190%7c180%7c192%7c195%7c175%7c200%7c205%7c202&start=0&sz=62"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}
```

```python
page=requests.get(link,headers=headers)
```

```python
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
```

```python
names = soup.find_all('div',class_='product-tile__name')
```

```python
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text().strip())
cust_name
```

```python
price = soup.find_all('div',class_='price')
price
```

```python
product_price = []
for i in range(0,len(price)):
    product_price.append(price[i].get_text().strip())
product_price
```

```python
product_link=[]
anchor=soup.find_all('a',class_='product-tile__image')
all_links=set()

for link in anchor:
    if(link.get('href') != '#'):
        product_link="https://www.oneill.com/fr/fr/t-shirts?cgid=men-tees&prefn1=published&prefv1=true&prefn2=seasons&prefv2=NOS%7c182%7c185%7c190%7c180%7c192%7c195%7c175%7c200%7c205%7c202&start=0&sz=62" +link.get('href')
        all_links.add(product_link)
        print(product_link)
```

```python
cust_name
product_price
product_link
```

```python
import pandas as pd
df = pd.DataFrame()
df
```

```python
df['Product_Name']=cust_name
df['Product_Price']=product_price
df['Product_Link']=product_link
```
