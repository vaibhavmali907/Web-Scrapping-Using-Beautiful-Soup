```python
from bs4 import BeautifulSoup as bs
import urllib.request
import requests
import re
```

```python
link="https://www.blue-tomato.com/de-DE/products/categories/Snowboard+Shop-00000000/gender/boys--girls--men--women/"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}
```

```python
page=requests.get(link,headers=headers)
```

```python
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
```

```python
names = soup.find_all('div',class_='ellipsis')
```

```python
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text().strip())
cust_name
```

```python
brand = soup.find_all('div',attrs={'class':'ellipsis'})
```

```python
product_brand = []
for i in range(0,len(brand)):
    product_brand.append(brand[i].get_text().strip())
product_brand
```

```python
product_brand
```

```python
price = soup.find_all('span',class_='price')
price
```

```python
product_price = []
for i in range(0,len(price)):
    product_price.append(price[i].get_text().strip())
product_price
```

```python
product_link=[]
anchor=soup.find_all('a',class_='name track-click track-load-producttile')
all_links=set()

for link in anchor:
    if(link.get('href') != '#'):
        product_link="https://www.blue-tomato.com/de-DE/products/categories/Snowboard+Shop-00000000/gender/boys--girls--men--women/" +link.get('href')
        all_links.add(product_link)
        print(product_link)
```

```python
html = urllib.request.urlopen('https://www.blue-tomato.com/de-DE/products/categories/Snowboard+Shop-00000000/gender/boys--girls--men--women/')
bs = bs(html, 'html.parser')
images = bs.find_all('img', {'src':re.compile('.jpg')})
for image in images: 
    print(image['src']+'\n')
```
